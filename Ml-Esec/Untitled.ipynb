```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
data = pd.read_csv("ToyotaCorolla (2).csv")
x = data['KM']
y = data['Weight']
z = data['Price']
```

1. Scatter Plot

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scatter plot")
plt.show()
```

2. Contour Plot

```python
plt.tricontourf(x,y,z,levels=20,cmap='jet')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Contour Plot")
plt.show()
```

3. 3d Surface Plot

```python
x = data['KM']
y = data['Doors']
z = data['Price']
```

```python
ax=plt.axes(projection='3d')
ax.set_title('3D Surface Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot_trisurf(x,y,z,cmap='jet')
plt.show()
```

4. Heat Map

```python
sns.heatmap(data[["Price","KM","Doors", "Weight"]].corr(),cmap='jet')
plt.show()
```

5. Box Plot

```python
plt.boxplot([data["Price"],data["HP"],data["KM"]])
plt.xticks([1,2,3],["Price","HP","KM"])
plt.show()
```
